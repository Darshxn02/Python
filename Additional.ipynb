```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 11]

# Plot
plt.plot(x, y, label='Line 1', color='blue', marker='o')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot Example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [5, 7, 6, 8, 7]

# Plot
plt.scatter(x, y, color='red')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot Example')
plt.show()
```

```python
# Data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]
color=['green','red','blue','black']
# Plot
plt.pie(sizes)
plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio makes the pie round
plt.show()
```

```python
# Data
Class = ['A', 'B', 'C', 'D']
values = [30, 25, 40, 35]
color=['green','red','blue','black']

# Plot
plt.bar(Class, values,color=color)
plt.xlabel('Class')
plt.ylabel('Numbe of students')
plt.title('Student in each class')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "sales": [100, 120, 130, 90, 160]
})

# Create a line plot
sns.lineplot(x="day", y="sales", data=data)

# Add titles and labels
plt.title("Sales Over Days")
plt.xlabel("Day")
plt.ylabel("Sales")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = ["Apple", "Banana", "Apple", "Orange", "Banana", "Banana", "Apple"]

# Create the count plot
sns.countplot(x=data)

# Add title and labels
plt.title("Fruit Counts")
plt.xlabel("Fruit")
plt.ylabel("Count")

# Show plot
plt.show()
```
